```python
# Titanic Dataset Exploratory Data Analysis

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Task 1: Load the Titanic dataset
# TODO: Load the Titanic dataset from the following URL
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
# df = ...

# Completed code for Task 1
df = pd.read_csv('/content/Titanic-Dataset.csv')
print("Titanic dataset loaded", df.head())

# Task 2: Inspect the data
# TODO: Display the first 5 rows of the dataset
# ...
#print(df.head())

# TODO: Display the last 5 rows of the dataset
# ...
#print(df.tail())

# TODO: Display the data types of each column
# ...
#print(df.dtypes)

# TODO: Display summary statistics of the numerical columns
# ...
#print(df.describe())

# Basic shape and concise info
print("Shape (rows, cols):", df.shape)
#df.info()

# Missing values per column
#df.isna().sum()

# Task 3: Clean and preprocess the data

# TODO: Create a new categorical feature 'PclassCategory' from 'Pclass'
# df['PclassCategory'] = ...

# 1 = first class-p, 2 = second class-p, 3 = class - p

passenger_class_mapping = {1: 'First Class', 2: 'Second Class', 3: 'Third Class'}
df['PclassCategory'] = df['Pclass'].map(passenger_class_mapping)
print(passenger_class_mapping)

# TODO: Create a new categorical feature 'CabinCategory' from 'Cabin'
# df['CabinCategory'] = ...

# Known vs Unknown (based on whether Cabin value exists)
df['CabinCategory'] = np.where(df['Cabin'].notna(),'Known','Unknown')
print(df['CabinCategory'])

# TODO: Handle missing values in the 'Age' column (use mean age)
df['Age']=df['Age'].fillna(df['Age'].mean())
print(df['Age'])

# TODO: Create a 'Family' feature by combining 'SibSp' and 'Parch'
df['Family'] = df['SibSp'] + df['Parch']
print(df['Family'])

# Task 4: Identify specific data points

# TODO: Find passengers with the highest fare
# highest_fare_passengers = ...
max_fare = df['Fare'].max()
highest_fare_passengers = df.loc[df['Fare'] == max_fare, ['PassengerId','Name','Fare','Pclass','Sex','Age']]
print("Passengers with the highest fare:")
display(highest_fare_passengers)

# TODO: Find passengers with the largest family size
# largest_family_passengers = ...


max_family = df['Family'].max()
largest_family_passengers = df.loc[df['Family'] == max_family, ['PassengerId','Name','Family','SibSp','Parch','Pclass','Sex','Age']]
print("Passengers with the largest family:")
display(largest_family_passengers)

# Task 5: Demonstrate Pandas Series and DataFrame functionalities



# TODO: Create a Pandas Series of passenger names
# names_series = ...

names_series = df['Name']
print("Names Series:")
display(names_series)

# TODO: Create a DataFrame of passenger names and ages
# names_ages_df = ...

passenger_names_ages_df = df[['Name', 'Age']]
print("Passenger Names and Ages DataFrame:")
display(passenger_names_ages_df)

# TODO: Slice the DataFrame to show only passengers older than 50
# older_passengers = ...

older_passengers = df.loc[df['Age'] > 50, ['PassengerId','Name','Age','Sex','Pclass']]
display(older_passengers.head())

# Task 6: Introduce multi-indexing

# TODO: Create a multi-index DataFrame grouped by 'Sex' and 'Pclass'
# multi_index_df = ...

multi_index_df = (
    df.groupby(['Sex','Pclass'])
      .agg(count=('PassengerId','size'),
           survival_rate=('Survived','mean'),
           avg_age=('Age','mean'),
           avg_fare=('Fare','mean'))
      .sort_index()
)

multi_index_df.index.names = ['Sex','Pclass']
print("Multi-Index DataFrame:")
display(multi_index_df)

# Task 7: Visualizations

# TODO: Create a histogram of passenger ages
# plt.figure(figsize=(10, 6))

# plt.title('Histogram of Passenger Ages')
# plt.xlabel('Age')
# plt.ylabel('Count')
# plt.show()

sns.set(style="whitegrid")

# 1) Age distribution
plt.figure(figsize=(10,6))
sns.histplot(df['Age'], bins=30, kde=True, color='teal')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
#plt.show()

# TODO: Create a bar plot of survival rates by passenger class
# 2) Survival rate by Pclass (bar of means)

plt.figure(figsize=(7,5))
sns.barplot(x='Pclass', y='Survived', data=df, palette='viridis', estimator=np.mean)
plt.title('Average Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Survival Rate')
plt.ylim(0,1)
#plt.show()

# TODO: Create a heatmap of correlation between numerical features
# 3) Correlation heatmap (numeric only)
plt.figure(figsize=(12,10))
corr = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap (Numeric Features)')
#plt.show()


# TODO: Create a pair plot of 'Age', 'Fare', 'SibSp', and 'Parch', colored by 'Survived'
# 4) Create Pair Plot ---

sns.set(style="ticks")
pair_plot_cols = ['Age', 'Fare', 'SibSp', 'Parch','Survived'] # Select subset of columns

sns.pairplot(
    data=df[pair_plot_cols],
    vars=['Age', 'Fare', 'SibSp', 'Parch'],
    hue='Survived',
    palette='Set2',
    diag_kind='kde',
    corner=True
)
plt.suptitle("Pair Plot of Age, Fare, SibSp, and Parch Colored by Survival", y=1.02)
plt.show()

# Task 8: Analyze survival rates

# TODO: Calculate and display survival rates by sex
# survival_by_sex = ...
#1)  Calculate and display survival rates by Sex

survival_by_sex = (
    df.groupby('Sex')['Survived']
      .mean()
      .mul(100)
      .round(2)
      .to_frame('Survival_Rate(%)')
)
print("Survival Rate by Sex:")
display(survival_by_sex)

# TODO: Calculate and display survival rates by passenger class
# survival_by_class = ...
# 2) # Calculate and display survival rates by Passenger Class

survival_by_pclass = (
    df.groupby('Pclass')['Survived']
      .mean()
      .mul(100)
      .round(2)
      .to_frame('Survival_Rate(%)')
)
print("\nSurvival rate by Passenger Class:")
display(survival_by_pclass)

# Task 9: Analyze age distribution

# TODO: Create a box plot of age distribution by passenger class and sex
# plt.figure(figsize=(12, 6))
# ...
# plt.title('Age Distribution by Passenger Class and Sex')
# plt.show()

# 1) Analyze age distribution
plt.figure(figsize=(12, 6))

sns.boxplot(
    x='Pclass',
    y='Age',
    hue='Sex',
    data=df,
    palette='Set2'
)

plt.title('Age Distribution by Passenger Class and Sex')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.legend(title='Sex')
plt.show()

# Task 10: Analyze fare distribution

# TODO: Create a violin plot of fare distribution by passenger class and survival status
# plt.figure(figsize=(12, 6))
# ...
# plt.title('Fare Distribution by Passenger Class and Survival Status')
# plt.show()

#1) Analyze fare distribution

plt.figure(figsize=(12, 6))

sns.violinplot(
    x='Pclass',
    y='Fare',
    hue='Survived',
    data=df,
    split=True,
    palette='Set3'
)

plt.title('Fare Distribution by Passenger Class and Survival Status')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.legend(title='Survived', loc='upper right')
plt.show()


print("EDA tasks completed.")
```
